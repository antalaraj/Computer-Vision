```python
# =============================================================================
# Project: Semantic Segmentation using U-Net
# Description:
#   This project implements a full U-Net architecture for pixel-level
#   binary segmentation. It automatically loads the Oxford-IIIT Pet dataset
#   using TensorFlow Datasets (TFDS) and trains the model to segment
#   pets from the background.
#
# Key Features:
#   - Architecture: Classic U-Net with skip connections.
#   - Metric: Dice Coefficient Loss for handling class imbalance.
#   - Data Pipeline: Automated ETL using TensorFlow Datasets (TFDS).
#
# Author: Raj Antala
# =============================================================================

import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds

# -----------------------------------------------------------------------------
# [1] Model Architecture: U-Net
# -----------------------------------------------------------------------------
def double_conv_block(x, n_filters):
    """
    Applies two successive 3x3 convolutions, each followed by Batch Normalization
    and ReLU activation. This is the building block of the U-Net.
    """
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    return x

def downsample_block(x, n_filters):
    """
    Encoder Block: Applies convolution layers followed by Max Pooling
    to reduce spatial dimensions (downsampling).
    Returns both the feature map (f) for skip connection and the pooled output (p).
    """
    f = double_conv_block(x, n_filters)
    p = layers.MaxPool2D(2)(f)
    return f, p

def upsample_block(x, conv_features, n_filters):
    """
    Decoder Block: Upsamples the input using Conv2DTranspose and concatenates
    it with the corresponding feature map from the encoder (skip connection).
    """
    x = layers.Conv2DTranspose(n_filters, 3, 2, padding="same")(x)
    x = layers.concatenate([x, conv_features]) # Skip connection
    x = double_conv_block(x, n_filters)
    return x

def build_unet(input_shape=(128, 128, 1)):
    """
    Assembles the complete U-Net architecture.
    """
    inputs = Input(input_shape)

    # --- Encoder (Contracting Path) ---
    f1, p1 = downsample_block(inputs, 64)
    f2, p2 = downsample_block(p1, 128)
    f3, p3 = downsample_block(p2, 256)
    f4, p4 = downsample_block(p3, 512)

    # --- Bottleneck ---
    bottleneck = double_conv_block(p4, 1024)

    # --- Decoder (Expansive Path) ---
    u6 = upsample_block(bottleneck, f4, 512)
    u7 = upsample_block(u6, f3, 256)
    u8 = upsample_block(u7, f2, 128)
    u9 = upsample_block(u8, f1, 64)

    # --- Output Layer ---
    # Sigmoid activation outputs a probability (0-1) for each pixel
    outputs = layers.Conv2D(1, 1, padding="same", activation="sigmoid")(u9)
    return models.Model(inputs, outputs, name="U-Net")

def dice_loss(y_true, y_pred):
    """
    Dice Coefficient Loss: Optimized for medical/binary segmentation tasks.
    It measures the overlap between the predicted mask and the ground truth.
    """
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + 1) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + 1)

# -----------------------------------------------------------------------------
# [2] Data Pipeline
# -----------------------------------------------------------------------------
def load_and_preprocess_data():
    """
    Loads the Oxford-IIIT Pet dataset from TensorFlow Datasets.
    Applies preprocessing: Resizing, Grayscale conversion, and Normalization.
    """
    print("Initializing Data Pipeline...")
    print("Loading Oxford-IIIT Pet dataset (Version 4.0.0)...")

    # Load dataset with metadata
    dataset, info = tfds.load('oxford_iiit_pet:4.0.0', with_info=True)

    def preprocess(data_item):
        # 1. Resize Image and Mask to fixed size (128x128)
        image = tf.image.resize(data_item['image'], (128, 128))
        mask = tf.image.resize(data_item['segmentation_mask'], (128, 128))

        # 2. Convert to Grayscale (simplifies model complexity)
        image = tf.image.rgb_to_grayscale(image)

        # 3. Normalize pixel values to [0, 1]
        image = image / 255.0

        # 4. Binarize Mask
        # Original labels: 1=Pet, 2=Background, 3=Border.
        # We target the Pet (1) vs everything else (0).
        mask = tf.where(mask == 1, 1.0, 0.0)

        return image, mask

    # Optimize pipeline with batching and prefetching
    train_ds = dataset['train'].map(preprocess).batch(16).prefetch(tf.data.AUTOTUNE)
    val_ds = dataset['test'].map(preprocess).batch(16).prefetch(tf.data.AUTOTUNE)

    return train_ds, val_ds

# -----------------------------------------------------------------------------
# [3] Main Execution
# -----------------------------------------------------------------------------
if __name__ == "__main__":
    # 1. Initialize Data
    train_dataset, val_dataset = load_and_preprocess_data()
    print("Dataset loaded and preprocessed.")

    # 2. Build Model
    print("Building U-Net Model...")
    model = build_unet(input_shape=(128, 128, 1))
    model.summary()

    # 3. Compile
    model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss=dice_loss,
                  metrics=['accuracy'])

    # 4. Train
    print("Starting Training Process...")
    history = model.fit(train_dataset,
                        validation_data=val_dataset,
                        epochs=5,
                        verbose=1)

    # 5. Evaluate & Visualize
    print("Visualizing Validation Results...")
    for image_batch, mask_batch in val_dataset.take(1):
        # Run inference
        pred_batch = model.predict(image_batch)

        # Visualize first image in the batch
        plt.figure(figsize=(12, 4))

        # Original Image
        plt.subplot(1, 3, 1)
        plt.imshow(image_batch[0, :, :, 0], cmap='gray')
        plt.title("Input Image")

        # Ground Truth Label
        plt.subplot(1, 3, 2)
        plt.imshow(mask_batch[0, :, :, 0], cmap='gray')
        plt.title("Ground Truth Mask")

        # Model Prediction
        plt.subplot(1, 3, 3)
        plt.imshow(pred_batch[0, :, :, 0] > 0.5, cmap='gray')
        plt.title("Predicted Segmentation")

        plt.show()
        break

    print("Process completed successfully.")
```

```python
# =============================================================================
# Project: Brain Tumor Segmentation using Mini U-Net
# Description:
#   A lightweight "Mini" U-Net architecture implemented in TensorFlow/Keras.
#   Designed to demonstrate semantic segmentation on medical imaging (MRI)
#   with limited data. It intentionally overfits a single image to visualize
#   the network's learning capability.
#
# Architecture:
#   - Encoder: 2-level downsampling with Convolutional blocks
#   - Decoder: 2-level upsampling with Skip Connections
#   - Output: Sigmoid activation for binary mask generation
#
# Author: Raj Antala
# =============================================================================

import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# -----------------------------------------------------------------------------
# [1] Model Architecture: Mini U-Net
# -----------------------------------------------------------------------------
def conv_block(x, n_filters):
    """
    Standard Convolutional Block:
    Applies two 3x3 convolutions with ReLU activation.
    BatchNormalization is omitted for stability with batch_size=1.
    """
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    return x

def build_mini_unet(input_shape=(128, 128, 1)):
    """
    Builds the Mini U-Net model.
    A reduced-depth U-Net (2 levels deep) suitable for small datasets/demos.
    """
    inputs = Input(input_shape)

    # --- Encoder (Downsampling Path) ---
    # Extracts high-level features while reducing spatial dimensions
    f1 = conv_block(inputs, 32)
    p1 = layers.MaxPool2D(2)(f1)

    f2 = conv_block(p1, 64)
    p2 = layers.MaxPool2D(2)(f2)

    # --- Bottleneck (Bridge) ---
    # The deepest layer connecting Encoder and Decoder
    bottleneck = conv_block(p2, 128)

    # --- Decoder (Upsampling Path) ---
    # Reconstructs spatial resolution using features from Bottleneck and Skip Connections
    u2 = layers.Conv2DTranspose(64, 3, 2, padding="same")(bottleneck)
    u2 = layers.concatenate([u2, f2]) # Skip connection preserves spatial info
    u2 = conv_block(u2, 64)

    u1 = layers.Conv2DTranspose(32, 3, 2, padding="same")(u2)
    u1 = layers.concatenate([u1, f1]) # Skip connection
    u1 = conv_block(u1, 32)

    # --- Output Layer ---
    # Binary classification per pixel (0=Background, 1=Tumor)
    outputs = layers.Conv2D(1, 1, padding="same", activation="sigmoid")(u1)

    return models.Model(inputs, outputs, name="Mini-U-Net")

def dice_loss(y_true, y_pred):
    """
    Dice Loss Function:
    Optimized for segmentation tasks, especially with class imbalance
    (e.g., small tumor vs large background).
    """
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + 1) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + 1)

# -----------------------------------------------------------------------------
# [2] Data Preprocessing
# -----------------------------------------------------------------------------
def load_single_image(image_path):
    """
    Loads and preprocesses a single MRI image.
    Generates a synthetic ground truth mask using thresholding for demonstration.
    """
    if not os.path.exists(image_path):
        print(f"[Error] File not found: {image_path}")
        return None, None

    # Load image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (128, 128))

    # Generate synthetic mask (Thresholding bright tumor regions)
    # In a real medical pipeline, this mask would be provided by radiologists.
    _, mask = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # Normalize pixel values to [0, 1]
    img = img / 255.0
    img = np.expand_dims(img, axis=-1) # Add channel dim
    img = np.expand_dims(img, axis=0)  # Add batch dim

    mask = mask / 255.0
    mask = np.expand_dims(mask, axis=-1)
    mask = np.expand_dims(mask, axis=0)

    return img, mask

# -----------------------------------------------------------------------------
# [3] Main Execution
# -----------------------------------------------------------------------------
if __name__ == "__main__":
    # Input Configuration
    IMAGE_FILE = "my_image.jpg"

    print(f"Loading image: {IMAGE_FILE}...")
    x_train, y_train = load_single_image(IMAGE_FILE)

    if x_train is not None:
        print("Initializing Mini U-Net...")
        model = build_mini_unet(input_shape=(128, 128, 1))

        model.compile(optimizer=Adam(learning_rate=0.001),
                      loss=dice_loss,
                      metrics=['accuracy'])

        print("Starting training (Overfitting demonstration)...")
        # Training for 100 epochs to ensure the model memorizes the specific features
        history = model.fit(x_train, y_train, epochs=100, verbose=0)
        print("Training completed successfully.")

        # --- Visualization ---
        print("Generating prediction...")
        pred = model.predict(x_train)

        plt.figure(figsize=(12, 4))

        # Original MRI
        plt.subplot(1, 3, 1)
        plt.imshow(x_train[0,:,:,0], cmap='gray')
        plt.title("Original MRI Scan")

        # Ground Truth Mask
        plt.subplot(1, 3, 2)
        plt.imshow(y_train[0,:,:,0], cmap='gray')
        plt.title("Target Mask (Ground Truth)")

        # Model Prediction
        plt.subplot(1, 3, 3)
        plt.imshow(pred[0,:,:,0] > 0.5, cmap='gray')
        plt.title("AI Prediction Output")

        plt.show()
```
